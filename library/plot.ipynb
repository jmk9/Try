```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
len_ls = [1, 2, 3]
```

```python
plt.title('line and color')
plt.plot(len_ls, [1, 1, 1], 'g-', label='solid')
plt.plot(len_ls, [2, 2, 2], 'r--', label='dashed')
plt.plot(len_ls, [3, 3, 3], 'b-.', label='dash-dot')
plt.plot(len_ls, [4, 4, 4], ':', color='violet', label='dotted')
plt.axis([1, 3, 0, 5])
plt.legend(loc='upper center', ncol=4)
plt.show()
```

```python
epoch = 100
epochs = np.linspace(0, epoch)
acc = epochs**2 + 20
```

```python
plt.subplot(121)
plt.title('accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.plot(epochs, acc, 'r-') #x범위, y범위, 그래프모양

plt.subplot(122)
plt.title('loss')
plt.xlabel('epoch')
plt.show()
```

```python
def graph_att(epoch, dim):
    epochs = np.linspace(0, epoch)
    
    if dim == '2':
        acc = epochs**2
    elif dim == '3':
        acc = -epochs**3
    
    return epochs, acc
```

```python
epc1, acc1 = graph_att(100, '2')
epc2, acc2 = graph_att(200, '3')
```

```python
plt.subplot(121)
plt.title('acc')
plt.plot(epc1, acc1, 'r-') #x범위, y범위, 그래프모양

plt.subplot(122)
plt.title('loss')
plt.plot(epc2, acc2, 'g-')
plt.show()
```

```python
plt.subplot(121)
plt.title('acc')
plt.plot(epc1, acc1, 'r-', label='accuracy')
plt.legend()

plt.subplot(122)
plt.title('loss')
plt.plot(epc2, acc2, 'g:', label='loss')
plt.legend()
plt.show()
```
